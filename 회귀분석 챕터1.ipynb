```python
# Loading Dataset
from sklearn.datasets import fetch_california_housing
from sklearn.datasets import load_boston
boston = load_boston()
california = fetch_california_housing()
```

```python
# Importing Packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.mlab as mlab
import math

%matplotlib inline
```

```python
# Defining Dataset
dataset = pd.DataFrame(boston.data, columns=boston.feature_names)
dataset['target'] = boston.target

# Dataset Explaination
# CRIM - 범죄발생율, ZN - 25000스퀘어 당 주거지역 비율, INDUS - 비유통산업 비율
# CHAS - 강을 끼고 있는가
# NOX - 질산포화도, RM - 거주자의 평균 방 갯수, AGE - 소유자의 나이, DIS - 직원 5명당 가중거리
# RAD - 고속도로 접근성, TAX - 1만 달러 당 세금비율, PTRATIO - 학생/교사비율
# B - 흑인의 비율, LSTAT - 잘 못사는 가구 비율
# Target - $1,000으로 환산 / 평당 가격
```

```python
# Regression

import statsmodels.api as sm
import statsmodels.formula.api as smf

linear_regression_api = smf.ols(formula='target ~ LSTAT', data=dataset)
fitted_model = linear_regression_api.fit()
fitted_model.summary()
```

```python
# Predict
def standardize(x):
    return(x-np.mean(x))/np.std(x)

fitted_values = fitted_model.predict()
original_values = dataset['target']

residuals = original_values - fitted_values
normalized_residuals = standardize(residuals)

# Graph
residual_scatter_plot = plt.scatter(dataset['RM'], normalized_residuals)
mean_residual = plt.plot([3,9], [0,0], '-', color='red', linewidth=2)

plt.show()
```
